```python
# üèÜ –†–ï–ô–¢–ò–ù–ì–ò –ó–ê –†–Ü–ó–ù–ò–ú–ò –ö–†–ò–¢–ï–†–Ü–Ø–ú–ò
print("üèÜ –ê–ù–ê–õ–Ü–ó –ù–ê–ô–ö–†–ê–©–ò–• –ú–û–î–ï–õ–ï–ô –ó–ê –ö–†–ò–¢–ï–†–Ü–Ø–ú–ò")
print("=" * 80)

# 1. –ù–ê–ô–ö–†–ê–©–ê –¢–û–ß–ù–Ü–°–¢–¨ –¶–Ü–ù–û–í–û–ì–û –ü–†–û–ì–ù–û–ó–£–í–ê–ù–ù–Ø (R¬≤ Score)
print("\nüìà 1. –¢–û–ß–ù–Ü–°–¢–¨ –¶–Ü–ù–û–í–û–ì–û –ü–†–û–ì–ù–û–ó–£–í–ê–ù–ù–Ø (R¬≤ Score):")
print("-" * 60)
regression_models = df_comparison[df_comparison['r2_score'].notna()].copy()
regression_models_sorted = regression_models.sort_values('r2_score', ascending=False)

for i, (model, row) in enumerate(regression_models_sorted.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    print(f"{medal} {model:30} | R¬≤: {row['r2_score']:.4f} ({row['r2_score']*100:.2f}%) | MAE: ${row['mae_dollars']:.2f}")

# 2. –ù–ê–ô–ö–†–ê–©–ê –¢–û–ß–ù–Ü–°–¢–¨ –ù–ê–ü–†–Ø–ú–ö–£ –†–£–•–£ (Direction Accuracy)
print("\nüéØ 2. –¢–û–ß–ù–Ü–°–¢–¨ –ù–ê–ü–†–Ø–ú–ö–£ –†–£–•–£ (Direction Accuracy):")
print("-" * 60)
direction_models = df_comparison[df_comparison['direction_accuracy'].notna()].copy()
direction_models_sorted = direction_models.sort_values('direction_accuracy', ascending=False)

for i, (model, row) in enumerate(direction_models_sorted.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    accuracy_pct = row['direction_accuracy'] * 100
    print(f"{medal} {model:30} | –¢–æ—á–Ω—ñ—Å—Ç—å: {accuracy_pct:.2f}% | –ß–∞—Å: {row['training_time_sec']:.3f}—Å")

# 3. –ù–ê–ô–ö–†–ê–©–ê –®–í–ò–î–ö–Ü–°–¢–¨ –¢–†–ï–ù–£–í–ê–ù–ù–Ø
print("\n‚ö° 3. –®–í–ò–î–ö–Ü–°–¢–¨ –¢–†–ï–ù–£–í–ê–ù–ù–Ø:")
print("-" * 60)
speed_models = df_comparison.copy()
speed_models_sorted = speed_models.sort_values('training_time_sec', ascending=True)

for i, (model, row) in enumerate(speed_models_sorted.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    time_desc = "–ú–∏—Ç—Ç—î–≤–æ" if row['training_time_sec'] == 0 else f"{row['training_time_sec']:.3f}—Å"
    print(f"{medal} {model:30} | –ß–∞—Å: {time_desc:>10} | –¢–∏–ø: {row['model_type']}")

# 4. –ù–ê–ô–ö–†–ê–©–ê –¢–û–ß–ù–Ü–°–¢–¨ –¶–Ü–ù–û–í–û–á –ü–û–ú–ò–õ–ö–ò (MAE)
print("\nüí∞ 4. –ù–ê–ô–ú–ï–ù–®–ê –¶–Ü–ù–û–í–ê –ü–û–ú–ò–õ–ö–ê (MAE in $):")
print("-" * 60)
mae_models = df_comparison[df_comparison['mae_dollars'].notna()].copy()
mae_models_sorted = mae_models.sort_values('mae_dollars', ascending=True)

for i, (model, row) in enumerate(mae_models_sorted.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    mape_text = f"MAPE: {row['mape_percent']:.2f}%" if pd.notna(row['mape_percent']) else "MAPE: N/A"
    print(f"{medal} {model:30} | MAE: ${row['mae_dollars']:>6.2f} | {mape_text}")

# 5. –ó–ê–ì–ê–õ–¨–ù–ò–ô –†–ï–ô–¢–ò–ù–ì (–∫–æ–º–ø–ª–µ–∫—Å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞)
print("\nüåü 5. –ó–ê–ì–ê–õ–¨–ù–ò–ô –†–ï–ô–¢–ò–ù–ì (–∫–æ–º–ø–ª–µ–∫—Å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞):")
print("-" * 60)

# –°—Ç–≤–æ—Ä—é—î–º–æ –∫–æ–º–ø–ª–µ–∫—Å–Ω—É –æ—Ü—ñ–Ω–∫—É –¥–ª—è –∫–æ–∂–Ω–æ—ó –º–æ–¥–µ–ª—ñ
def calculate_overall_score(row):
    score = 0
    
    # R¬≤ Score (–¥–ª—è —Ä–µ–≥—Ä–µ—Å—ñ–π–Ω–∏—Ö –º–æ–¥–µ–ª–µ–π) - –≤–∞–≥ 30%
    if pd.notna(row['r2_score']):
        score += row['r2_score'] * 0.3
    
    # Direction Accuracy (–¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ–π–Ω–∏—Ö) - –≤–∞–≥ 30%  
    if pd.notna(row['direction_accuracy']):
        score += row['direction_accuracy'] * 0.3
    
    # –®–≤–∏–¥–∫—ñ—Å—Ç—å —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è (—ñ–Ω–≤–µ—Ä—Ç–æ–≤–∞–Ω–∞, –Ω–æ—Ä–º–∞–ª—ñ–∑–æ–≤–∞–Ω–∞) - –≤–∞–≥ 20%
    max_time = df_comparison['training_time_sec'].max()
    if max_time > 0:
        speed_score = 1 - (row['training_time_sec'] / max_time)
        score += speed_score * 0.2
    
    # MAE (—ñ–Ω–≤–µ—Ä—Ç–æ–≤–∞–Ω–∞, –Ω–æ—Ä–º–∞–ª—ñ–∑–æ–≤–∞–Ω–∞) - –≤–∞–≥ 20%
    if pd.notna(row['mae_dollars']):
        max_mae = mae_models['mae_dollars'].max()
        min_mae = mae_models['mae_dollars'].min()
        if max_mae > min_mae:
            mae_score = 1 - ((row['mae_dollars'] - min_mae) / (max_mae - min_mae))
            score += mae_score * 0.2
    
    return score

df_comparison['overall_score'] = df_comparison.apply(calculate_overall_score, axis=1)
overall_sorted = df_comparison.sort_values('overall_score', ascending=False)

for i, (model, row) in enumerate(overall_sorted.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    score_pct = row['overall_score'] * 100
    print(f"{medal} {model:30} | –ó–∞–≥–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞: {score_pct:>5.1f}% | –¢–∏–ø: {row['model_type']}")

print(f"\n‚úÖ –ê–Ω–∞–ª—ñ–∑ —Ä–µ–π—Ç–∏–Ω–≥—ñ–≤ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
print(f"üèÜ –ó–∞–≥–∞–ª—å–Ω–∏–π –ø–µ—Ä–µ–º–æ–∂–µ—Ü—å: {overall_sorted.index[0]}")
print(f"üìä –ù–∞–π–∫—Ä–∞—â–∞ —Ä–µ–≥—Ä–µ—Å—ñ–π–Ω–∞ –º–æ–¥–µ–ª—å: {regression_models_sorted.index[0]}")
print(f"üéØ –ù–∞–π–∫—Ä–∞—â–∞ –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ–π–Ω–∞ –º–æ–¥–µ–ª—å: {direction_models_sorted.index[0]}")
print(f"‚ö° –ù–∞–π—à–≤–∏–¥—à–∞ –º–æ–¥–µ–ª—å: {speed_models_sorted.index[0]}")
```

```python
# üéØ –§–Ü–ù–ê–õ–¨–ù–Ü –í–ò–°–ù–û–í–ö–ò –¢–ê –†–ï–ö–û–ú–ï–ù–î–ê–¶–Ü–á
print("üéØ –§–Ü–ù–ê–õ–¨–ù–Ü –í–ò–°–ù–û–í–ö–ò –¢–ê –†–ï–ö–û–ú–ï–ù–î–ê–¶–Ü–á")
print("=" * 80)

# –û—Ç—Ä–∏–º—É—î–º–æ –Ω–∞–π–∫—Ä–∞—â—ñ –º–æ–¥–µ–ª—ñ –∑–∞ –∫–æ–∂–Ω–∏–º –∫—Ä–∏—Ç–µ—Ä—ñ—î–º
best_overall = overall_sorted.index[0]
best_regression = regression_models_sorted.index[0] if not regression_models_sorted.empty else "N/A"
best_classification = direction_models_sorted.index[0] if not direction_models_sorted.empty else "N/A"
fastest_model = speed_models_sorted.index[0]
most_accurate_price = mae_models_sorted.index[0] if not mae_models_sorted.empty else "N/A"

print(f"\nüèÜ –¢–û–ü –†–ï–ó–£–õ–¨–¢–ê–¢–ò:")
print("-" * 60)
print(f"üåü –ó–∞–≥–∞–ª—å–Ω–∏–π –ø–µ—Ä–µ–º–æ–∂–µ—Ü—å:        {best_overall}")
print(f"üìà –ù–∞–π–∫—Ä–∞—â–∞ —Ä–µ–≥—Ä–µ—Å—ñ–π–Ω–∞ –º–æ–¥–µ–ª—å:  {best_regression}")
print(f"üéØ –ù–∞–π–∫—Ä–∞—â–∞ –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ–π–Ω–∞:     {best_classification}") 
print(f"üí∞ –ù–∞–π—Ç–æ—á–Ω—ñ—à–∞ –∑–∞ —Ü—ñ–Ω–æ—é:         {most_accurate_price}")
print(f"‚ö° –ù–∞–π—à–≤–∏–¥—à–∞ –º–æ–¥–µ–ª—å:            {fastest_model}")

print(f"\nüìä –ö–õ–Æ–ß–û–í–Ü –Ü–ù–°–ê–ô–¢–ò:")
print("-" * 60)

# –ê–Ω–∞–ª—ñ–∑ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—ñ —Ç–∏–ø—ñ–≤ –º–æ–¥–µ–ª–µ–π
type_performance = df_comparison.groupby('model_type').agg({
    'overall_score': 'mean',
    'training_time_sec': 'mean', 
    'r2_score': 'mean',
    'direction_accuracy': 'mean'
}).round(3)

print("1. üìà –ï–§–ï–ö–¢–ò–í–ù–Ü–°–¢–¨ –¢–ò–ü–Ü–í –ú–û–î–ï–õ–ï–ô:")
for model_type, scores in type_performance.iterrows():
    score = scores['overall_score']
    time = scores['training_time_sec']
    print(f"   ‚Ä¢ {model_type:20}: –û—Ü—ñ–Ω–∫–∞ {score:.3f}, –ß–∞—Å {time:.2f}—Å")

print(f"\n2. üéØ –ù–ê–ô–í–ê–ñ–õ–ò–í–Ü–®–Ü –ó–ù–ê–•–Ü–î–ö–ò:")

# Foundation Model –∞–Ω–∞–ª—ñ–∑
sundial_row = df_comparison.loc['üåÖ Sundial Foundation Model']
print(f"   üåÖ SUNDIAL FOUNDATION MODEL:")
print(f"      ‚Ä¢ –ù–∞–π–∫—Ä–∞—â–∞ —Ç–æ—á–Ω—ñ—Å—Ç—å –Ω–∞–ø—Ä—è–º–∫—É: {sundial_row['direction_accuracy']*100:.1f}%")
print(f"      ‚Ä¢ –ß—É–¥–æ–≤–∞ —Ü—ñ–Ω–æ–≤–∞ —Ç–æ—á–Ω—ñ—Å—Ç—å: MAPE {sundial_row['mape_percent']:.2f}%")
print(f"      ‚Ä¢ Zero-shot –Ω–∞–≤—á–∞–Ω–Ω—è: 0 —Å–µ–∫—É–Ω–¥ —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è")
print(f"      ‚Ä¢ –ó–∞–≥–∞–ª—å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞: {sundial_row['overall_score']*100:.1f}%")

# Linear Regression –∞–Ω–∞–ª—ñ–∑  
lr_row = df_comparison.loc['Linear Regression']
print(f"\n   üìà LINEAR REGRESSION:")
print(f"      ‚Ä¢ –§–µ–Ω–æ–º–µ–Ω–∞–ª—å–Ω–∏–π R¬≤: {lr_row['r2_score']*100:.2f}%")
print(f"      ‚Ä¢ –ú—ñ–Ω—ñ–º–∞–ª—å–Ω–∞ –ø–æ–º–∏–ª–∫–∞: ${lr_row['mae_dollars']:.2f}")
print(f"      ‚Ä¢ –ù–∞–¥—à–≤–∏–¥–∫—ñ—Å—Ç—å: {lr_row['training_time_sec']:.3f}—Å")

# AutoML –∞–Ω–∞–ª—ñ–∑
automl_row = df_comparison.loc['AutoML Ready Model']
print(f"\n   ü§ñ AUTOML READY MODEL:")
print(f"      ‚Ä¢ –ù–∞–π–∫—Ä–∞—â–∞ –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è: {automl_row['direction_accuracy']*100:.2f}%")
print(f"      ‚Ä¢ –®–≤–∏–¥–∫–µ –≤–ø—Ä–æ–≤–∞–¥–∂–µ–Ω–Ω—è: {automl_row['training_time_sec']:.3f}—Å")
print(f"      ‚Ä¢ –ù–µ –ø–æ—Ç—Ä–µ–±—É—î –µ–∫—Å–ø–µ—Ä—Ç–∏–∑–∏")

print(f"\n3. üí° –ü–†–ê–ö–¢–ò–ß–ù–Ü –†–ï–ö–û–ú–ï–ù–î–ê–¶–Ü–á:")
print("-" * 60)

print("üéØ –î–õ–Ø –†–Ü–ó–ù–ò–• –°–¶–ï–ù–ê–†–Ü–á–í –í–ò–ö–û–†–ò–°–¢–ê–ù–ù–Ø:")

print(f"\n   üöÄ –î–õ–Ø –®–í–ò–î–ö–û–ì–û –ü–†–û–¢–û–¢–ò–ü–£–í–ê–ù–ù–Ø:")
print(f"      ‚úÖ {fastest_model}")
print(f"      üìä –ü—Ä–∏—á–∏–Ω–∞: {speed_models_sorted.loc[fastest_model, 'training_time_sec']:.4f}—Å —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è")

print(f"\n   üìà –î–õ–Ø –¢–û–ß–ù–û–ì–û –ü–†–û–ì–ù–û–ó–£–í–ê–ù–ù–Ø –¶–Ü–ù:")
print(f"      ‚úÖ {best_regression}")
print(f"      üìä –ü—Ä–∏—á–∏–Ω–∞: R¬≤ = {regression_models_sorted.loc[best_regression, 'r2_score']*100:.2f}%")

print(f"\n   üéØ –î–õ–Ø –¢–û–†–ì–û–í–ò–• –°–ò–ì–ù–ê–õ–Ü–í:")
print(f"      ‚úÖ {best_classification}")
print(f"      üìä –ü—Ä–∏—á–∏–Ω–∞: {direction_models_sorted.loc[best_classification, 'direction_accuracy']*100:.2f}% —Ç–æ—á–Ω—ñ—Å—Ç—å –Ω–∞–ø—Ä—è–º–∫—É")

print(f"\n   üè¢ –î–õ–Ø PRODUCTION –°–ò–°–¢–ï–ú–ò:")
print(f"      ‚úÖ üåÖ Sundial Foundation Model")
print(f"      üìä –ü—Ä–∏—á–∏–Ω–∞: Zero-shot + 75% —Ç–æ—á–Ω—ñ—Å—Ç—å + MAPE 3.95%")

print(f"\n4. üîÆ –ú–ê–ô–ë–£–¢–ù–Ü –ù–ê–ü–†–Ø–ú–ö–ò:")
print("-" * 60)
print("   üß† Fine-tune Sundial –Ω–∞ —Å–ø–µ—Ü–∏—Ñ—ñ—á–Ω–∏—Ö —Ñ—ñ–Ω–∞–Ω—Å–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö")
print("   üîÑ Ensemble –∑ Linear Regression + Sundial")  
print("   üìä A/B —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è –Ω–∞ –∂–∏–≤–∏—Ö —Ä–∏–Ω–∫–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö")
print("   ‚ö° –û–ø—Ç–∏–º—ñ–∑–∞—Ü—ñ—è —à–≤–∏–¥–∫–æ—Å—Ç—ñ –¥–ª—è HFT")
print("   üåê –Ü–Ω—Ç–µ–≥—Ä–∞—Ü—ñ—è –∑ real-time data streams")

print(f"\n‚úÖ –ü–Ü–î–°–£–ú–û–ö:")
print("=" * 80)
print("üèÜ –ü—Ä–æ–µ–∫—Ç –¥–µ–º–æ–Ω—Å—Ç—Ä—É—î –ø–æ–≤–Ω–∏–π —Å–ø–µ–∫—Ç—Ä ML –ø—ñ–¥—Ö–æ–¥—ñ–≤")
print("üìä –û—Ç 49% –¥–æ 98% —Ç–æ—á–Ω–æ—Å—Ç—ñ –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –∑–∞–≤–¥–∞–Ω–Ω—è")  
print("‚ö° –í—ñ–¥ –º–∏—Ç—Ç—î–≤–æ–≥–æ –¥–æ 2 —Ö–≤–∏–ª–∏–Ω —á–∞—Å—É —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è")
print("üéØ –ö–æ–∂–Ω–∞ –º–æ–¥–µ–ª—å –º–∞—î —Å–≤–æ—é –Ω—ñ—à—É –∑–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è")
print("üåÖ Sundial Foundation Model - –º–∞–π–±—É—Ç–Ω—î —Ñ—ñ–Ω–∞–Ω—Å–æ–≤–æ–≥–æ ML")

print(f"\nüéì –ù–ê–í–ß–ê–õ–¨–ù–ê –¶–Ü–ù–ù–Ü–°–¢–¨:")
print("üìö –û–ø–∞–Ω–æ–≤–∞–Ω–æ 6 —Ä—ñ–∑–Ω–∏—Ö —Ç–∏–ø—ñ–≤ ML –º–æ–¥–µ–ª–µ–π")
print("üîß –í—ñ–¥ –∫–ª–∞—Å–∏—á–Ω–∏—Ö –¥–æ foundation models")
print("üìà –ö–æ–º–ø–ª–µ–∫—Å–Ω–∞ –æ—Ü—ñ–Ω–∫–∞ —Ç–∞ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è")
print("üíº –ì–æ—Ç–æ–≤—ñ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó –¥–ª—è —Ä–µ–∞–ª—å–Ω–∏—Ö –ø—Ä–æ–µ–∫—Ç—ñ–≤")

print("\n" + "üåü" * 40)
print("–§—ñ–Ω–∞–ª—å–Ω–µ –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –º–æ–¥–µ–ª–µ–π –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
print("üåü" * 40)
```

```python
# üíæ –ï–ö–°–ü–û–†–¢ –†–ï–ó–£–õ–¨–¢–ê–¢–Ü–í –¢–ê –†–ï–ó–Æ–ú–ï
from datetime import datetime

print("üíæ –ó–ë–ï–†–ï–ñ–ï–ù–ù–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–Ü–í...")

# –ï–∫—Å–ø–æ—Ä—Ç –ø–æ–≤–Ω–æ—ó —Ç–∞–±–ª–∏—Ü—ñ
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
results_file = f"final_models_comparison_{timestamp}.csv"

# –ü—ñ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–∏—Ö –¥–ª—è –µ–∫—Å–ø–æ—Ä—Ç—É –∑ –¥–æ–¥–∞—Ç–∫–æ–≤–∏–º–∏ –∫–æ–ª–æ–Ω–∫–∞–º–∏
export_df = df_comparison.copy()
export_df['rank_overall'] = export_df['overall_score'].rank(ascending=False, method='min').astype(int)
export_df['rank_regression'] = export_df['r2_score'].rank(ascending=False, method='min')
export_df['rank_classification'] = export_df['direction_accuracy'].rank(ascending=False, method='min')
export_df['rank_speed'] = export_df['training_time_sec'].rank(ascending=True, method='min').astype(int)

# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –¥–ª—è —á–∏—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—ñ
export_df['r2_score_percent'] = (export_df['r2_score'] * 100).round(2)
export_df['direction_accuracy_percent'] = (export_df['direction_accuracy'] * 100).round(2)

# –ü–µ—Ä–µ–≤–ø–æ—Ä—è–¥–∫–æ–≤—É—î–º–æ –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è –∑—Ä—É—á–Ω–æ—Å—Ç—ñ
columns_order = [
    'notebook', 'model_type', 'task_type',
    'rank_overall', 'overall_score',
    'rank_regression', 'r2_score', 'r2_score_percent',
    'rank_classification', 'direction_accuracy', 'direction_accuracy_percent', 
    'mae_dollars', 'mse', 'mape_percent',
    'rank_speed', 'training_time_sec',
    'complexity', 'interpretability'
]

export_df_final = export_df[columns_order].round(4)

# –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –≤ CSV
export_df_final.to_csv(results_file)

print(f"‚úÖ –†–µ–∑—É–ª—å—Ç–∞—Ç–∏ –∑–±–µ—Ä–µ–∂–µ–Ω–æ –≤: {results_file}")

# –°—Ç–≤–æ—Ä—é—î–º–æ –∫–æ—Ä–æ—Ç–∫–∏–π —Ä–µ–∑—é–º–µ —Ç–æ–ø-5 –º–æ–¥–µ–ª–µ–π
print(f"\nüìã –ö–û–†–û–¢–ö–ò–ô –†–ï–ó–Æ–ú–ï –¢–û–ü-5 –ú–û–î–ï–õ–ï–ô:")
print("=" * 80)

top5_models = overall_sorted.head(5)
summary_table = []

for i, (model_name, row) in enumerate(top5_models.iterrows(), 1):
    medal = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i}."
    
    # –û—Å–Ω–æ–≤–Ω–∞ –º–µ—Ç—Ä–∏–∫–∞ (R¬≤ –∞–±–æ Direction Accuracy)
    main_metric = ""
    if pd.notna(row['r2_score']):
        main_metric = f"R¬≤: {row['r2_score']*100:.1f}%"
    elif pd.notna(row['direction_accuracy']):
        main_metric = f"Dir: {row['direction_accuracy']*100:.1f}%"
    
    # MAE —è–∫—â–æ —î
    mae_info = f"MAE: ${row['mae_dollars']:.2f}" if pd.notna(row['mae_dollars']) else ""
    
    # –ß–∞—Å —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è
    time_info = "–ú–∏—Ç—Ç—î–≤–æ" if row['training_time_sec'] == 0 else f"{row['training_time_sec']:.3f}—Å"
    
    summary_table.append({
        'Rank': medal,
        'Model': model_name,
        'Score': f"{row['overall_score']*100:.1f}%",
        'Main_Metric': main_metric,
        'MAE': mae_info,
        'Time': time_info,
        'Type': row['model_type']
    })
    
    print(f"{medal} {model_name:30} | –û—Ü—ñ–Ω–∫–∞: {row['overall_score']*100:5.1f}% | {main_metric:12} | {time_info:>8} | {row['model_type']}")

# –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –∫–æ—Ä–æ—Ç–∫–∏–π —Ä–µ–∑—é–º–µ
summary_df = pd.DataFrame(summary_table)
summary_file = f"top5_models_summary_{timestamp}.csv"
summary_df.to_csv(summary_file, index=False)

print(f"\nüíæ –¢–æ–ø-5 —Ä–µ–∑—é–º–µ –∑–±–µ—Ä–µ–∂–µ–Ω–æ –≤: {summary_file}")

# –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ —Ç–∏–ø–∞—Ö –º–æ–¥–µ–ª–µ–π
print(f"\nüìä –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ü–û –¢–ò–ü–ê–• –ú–û–î–ï–õ–ï–ô:")
print("-" * 60)

type_stats = df_comparison.groupby('model_type').agg({
    'overall_score': ['count', 'mean', 'max'],
    'training_time_sec': 'mean'
}).round(3)

type_stats.columns = ['Count', 'Avg_Score', 'Max_Score', 'Avg_Time']
type_stats = type_stats.sort_values('Max_Score', ascending=False)

for model_type, stats in type_stats.iterrows():
    print(f"{model_type:20} | –ö—ñ–ª—å–∫—ñ—Å—Ç—å: {stats['Count']:1.0f} | –°–µ—Ä.–æ—Ü—ñ–Ω–∫–∞: {stats['Avg_Score']:.3f} | –ú–∞–∫—Å: {stats['Max_Score']:.3f} | –ß–∞—Å: {stats['Avg_Time']:.2f}—Å")

# –ó–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Ç–∏–ø—ñ–≤
type_stats_file = f"model_types_stats_{timestamp}.csv"
type_stats.to_csv(type_stats_file)

print(f"\nüíæ –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ç–∏–ø—ñ–≤ –∑–±–µ—Ä–µ–∂–µ–Ω–∞ –≤: {type_stats_file}")

print(f"\nüéØ –§–ê–ô–õ–ò –î–õ–Ø –ê–ù–ê–õ–Ü–ó–£:")
print(f"   üìä –ü–æ–≤–Ω—ñ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏: {results_file}")
print(f"   üèÜ –¢–æ–ø-5 –º–æ–¥–µ–ª–µ–π:    {summary_file}")
print(f"   üìà –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ç–∏–ø—ñ–≤: {type_stats_file}")

print(f"\n‚úÖ –ï–∫—Å–ø–æ—Ä—Ç –∑–∞–≤–µ—Ä—à–µ–Ω–æ! –í—Å—ñ —Ñ–∞–π–ª–∏ –≥–æ—Ç–æ–≤—ñ –¥–ª—è –ø–æ–¥–∞–ª—å—à–æ–≥–æ –∞–Ω–∞–ª—ñ–∑—É.")

# –ü–æ–∫–∞–∑—É—î–º–æ —Ñ—ñ–Ω–∞–ª—å–Ω—É —Ç–∞–±–ª–∏—Ü—é –¥–ª—è notebook
print(f"\nüìã –§–Ü–ù–ê–õ–¨–ù–ê –ó–í–ï–î–ï–ù–ê –¢–ê–ë–õ–ò–¶–Ø:")
display(export_df_final.head(10))
```
